```python
def explore(x):
    x=int(x)
    column=col[x]
    print('column name: ',column)
    print(f'number of unique values in "{column}":',data[column].nunique())
    count=data[column].nunique()
    print(f'Count of unique values:\n',data[column].value_counts())
    if count<10:
        ax=sns.countplot(data[column],hue=data[col[0]])
        plt.xticks(rotation=90)

        for p in ax.patches:
            ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.01, p.get_height()+0.1))
    
    else:
        plt.figure(figsize=(10,10))
        ax=sns.countplot(y=data[column],hue=data[col[0]])

```
